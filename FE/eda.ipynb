```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("raw.csv",header=1)
df.columns = [str(f).strip() for f in df.columns]
df.head()
```

```python
df[df.isnull().any(axis=1)]
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
df=df.dropna().reset_index(drop=True)

df.head()
```

```python
df.info()
columns_to_convert = ['day', 'month', 'year', 'Temperature', 'RH', 'Ws','Rain', 'FFMC',
       'DMC', 'DC', 'ISI', 'BUI', 'FWI', 'Region']

try:
    df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
    print("Conversion successful!")
except ValueError as e:
    print("Error occurred during conversion:", str(e))
```

```python
## Let ave the cleaned dataset
df.to_csv('final.csv',index=False)
```
